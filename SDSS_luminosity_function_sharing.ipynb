# The Cosmic Web
Galaxies on the near Universe and how they are influenced by the Cosmic web.
 
This document presents an analysis on Geometrical environmens of Galaxies in the SDSS Survey and their implication to their physical properties.


In particular, in this notebook we explore the Luminosity funciont, which is the number density of galaxies per volume per magnitude.

This is fitted with a Schechter profile (see "thesis_functions_lum.py") using the extremization of the log-likelihood method. 

Relevant Parameters to be choose are found under "Basics.Reading Data" Subsection.
- sub_pop, Choose according to relevant subpopulation under study
- key_dens, Choose according to relevant density tracer.

First relevant result, the graphs here show the disparities among environments once normalized by the whole set behavior and controled for 0-th order density effect (avegare density per environment)

This is similar to other results in the literature as expected. The question is, these disparities remain once the effect of density is controlled?

##### Results relevance: Information Criteria

```python
def AIC(y, yerr,k,model):
    sigma2 = yerr**2
    ll = -0.5*np.sum((y-model)**2/sigma2 + np.log(sigma2))
    return -2*ll + k*2


def BIC(y, yerr,k,model):
    sigma2 = yerr**2
    n = len(y)
    ll = -0.5*np.sum((y-model)**2/sigma2 + np.log(sigma2))
    return -2*ll + k*np.log(n)

def AICc(y, yerr,k,model):
    sigma2 = yerr**2
    n = len(y)
    ll = -0.5*np.sum((y-model)**2/sigma2 + np.log(sigma2))
    return -2*ll + k*2*(k+1)/(n-k-1)
```
